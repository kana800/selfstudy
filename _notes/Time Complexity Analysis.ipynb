### How to analyze the time complexity

The running time of the algorithm might depend some factors.
1. Single vs Multi processor
2. Read/Write Speed to Memory
3. 32 bit vs 64 bit 
4. Input

we only consider how the time taken by the program grows with the input given to the program. We are interested in the rate of growth of time taken with respect to the input.

## Asymptotic notations


## Big O Notation

Gives us the upper bound of the rate of growth of function

**$O(g(n))$ where $f(n)$ there exits contants *c* and $n_{0}$, $f(n) <= c*g(n)$, for $n >= n_{0}$**

## Omega Notation

Gives us the lower bound of the rate of growth of a function

**$O(g(n))$ where $f(n)$ there exits contants *c* and $n_{0}$, $f(n) >= c*g(n)$, for $n >= n_{0}$**

## Theta Notation

Gives us the tight bound of the rate of growth of a function

## Rules

We analyze time complexity for:

1. large numbers **n->inifinity**
2. Worst case scenario

### Rule 1

1. Drop lower order terms
2. Drop the constant multiplier

Example:

Algorithm 1: $T(n)=17n^{3}+3n^{2}+6n+8$ 

$n^{3}$ (n ->infinity) $O(n^{3})$

Algorithm 2: $T(n)=18n^4+17n^{3}+3n^{2}+6n+8$ 

$n^{4}$ (n ->infinity) $O(n^{4})$

Algorithm 3: $T(n)=6n+log(n)$ 

$n$ (n ->infinity) $O(n)$

### Rule 2

Running time is equal to the **summation** of running time of ll the fragments.

1. Simple Statements $O(1)$
```C
int a;
a = 5;
a++;
```
2. Single Loop $O(n)$

```C
for(int i = 0; i < n ; i++)
{
    //simple statements
}
```

3. Nested Loop $O(n^{2})$

```C
for(int i = 0; i < n ; i++)
{
    for(int j = 0; j < n ; j++)
    {
        //simple statements
    }
```

Example:

```C
int a;
a = 5;
a++;
for(int i = 0; i < n ; i++)
{
    //simple statements
}

for(int i = 0; i < n ; i++)
{
    for(int j = 0; j < n ; j++)
    {
        //simple statements
    }
```
$T(n) = O(1) + O(n) + O(n^{2}) = O(n^{2})$ 

---
### Rule 3

Pick complexity of condition which is the worst case

```C
if (some condition)
{
    for(int i = 0; i < n ; i++)
    {
        //simple statements
    }
}
else 
{
   for(int i = 0; i < n ; i++)
    {
        for(int j = 0; j < n ; j++)
        {
            //simple statements
        } 
}
```

$T(n) = O(n^{2})$
